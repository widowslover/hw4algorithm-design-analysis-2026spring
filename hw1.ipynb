根据下列递归公式, 计算下列 $T(n)$ 对应的的渐近上界。要求所求的边界尽可能的紧（tight）, 请写明步骤。

1. $T(1)=1; T(n)=T(n/4)+1$ for $n>1$
2. $T(1)=1;T(n)=3T(n/3)+n^2$ for $n>1$
3. $T(1)=1;T(n)=T(2n/3)+1$ for $n>1$
4. $T(1)=1;T(n)=5T(n/4)+n$ for $n>1$
5. $T(n)=1 \ for\ n \le 2 ; T(n)=T(\sqrt{n})+1 \ for \ n>2$


1. $T(n)=T(n/4)+1$  
   展开 $k$ 层后规模为 $n/4^k$，令 $n/4^k=1$，得 $k=\log_4 n$。每层代价为常数，所以：
   $$T(n)=\Theta(\log n)$$

2. $T(n)=3T(n/3)+n^2$  
   主定理中 $a=3,b=3,f(n)=n^2$，有
   $$n^{\log_b a}=n^{\log_3 3}=n$$
   因为 $n^2$ 比 $n$ 多一个多项式因子，所以：
   $$T(n)=\Theta(n^2)$$

3. $T(n)=T(2n/3)+1$  
   展开 $k$ 层后规模为 $(2/3)^k n$，令 $(2/3)^k n\le 1$，得 $k=\Theta(\log n)$。每层代价为常数，所以：
   $$T(n)=\Theta(\log n)$$

4. $T(n)=5T(n/4)+n$  
   主定理中 $a=5,b=4,f(n)=n$，有
   $$n^{\log_b a}=n^{\log_4 5}\approx n^{1.16}$$
   $f(n)=n$ 比 $n^{\log_4 5}$ 小，因此：
   $$T(n)=\Theta(n^{\log_4 5})$$

5. $T(n)=T(\sqrt n)+1$  
   第 $k$ 层后的规模为 $n^{1/2^k}$。令 $n^{1/2^k}\le 2$，两边取对数：
   $$\frac{\log n}{2^k}\le \log 2$$
   所以 $2^k=\Omega(\log n)$，即 $k=\Theta(\log\log n)$。因此：
   $$T(n)=\Theta(\log\log n)$$

## 问题 3

给定一个包含n个元素的数组 `profits` , 它的第 `i` 个元素 `profits[i]` 表示一支股票第 `i` 天的**收益**（正数表示涨, 负数表示跌）。你只能选择 **某一天** 买入这只股票, 并选择在 **未来的某一个不同的日子** 卖出该股票。

1. 设计一个算法来计算你所能获取的最大利润和对应买入和卖出的日期。请分析算法方案, 计算其时间复杂度, 并且使用python编程实现该算法。

2. \* 设计一个时间复杂度为 $O(n)$的算法实现该算法

e.g. :
---
profits=[3,2,1,-7,5,2,-1,3,-1], 第5天买入, 第8天卖出, 收益最大：9


因为 `profits[i]` 表示第 $i$ 天的收益，所以若第 $i$ 天买入、第 $j$ 天卖出，且 $j>i$，总收益为：

$$profits[i]+profits[i+1]+\cdots+profits[j]$$
枚举所有买入日期 $i$ 和卖出日期 $j$，维护区间和，找出收益最大的方案。

算法步骤：

1. 固定买入日期 `i`；
2. 从 `i` 往后枚举卖出日期 `j`，一边累加收益；
3. 只在 `j > i` 时更新答案，保证买入和卖出是不同日期。

时间复杂度：$O(n^2)$。  
额外空间复杂度：$O(1)$。  


````python
def max_profit_bruteforce(profits):
    """返回 (最大收益, 买入日期, 卖出日期)，日期从 1 开始计数。"""
    n = len(profits)
    if n < 2:
        raise ValueError("至少需要 2 天数据")

    best_profit = float("-inf")
    best_buy = best_sell = None

    for i in range(n - 1):
        total = 0
        for j in range(i, n):
            total += profits[j]
            if j > i and total > best_profit:
                best_profit = total
                best_buy = i + 1
                best_sell = j + 1

    return best_profit, best_buy, best_sell

profits = [3, 2, 1, -7, 5, 2, -1, 3, -1]
ans = max_profit_bruteforce(profits)
print("profits =", profits)
print(f"最大收益：{ans[0]}，买入日期：第{ans[1]}天，卖出日期：第{ans[2]}天")
print("算法时间复杂度：O(n^2)，额外空间复杂度：O(1)")
````

## 问题 4

给定一个包含不同整数元素的数组 $ A[1..n]$ ,并且满足条件：$A[1]>A[2]$ 并且 $A[n-1]<A[n]$; 规定：如果一个元素比它两边的邻居元素都小, 即：$A[x]<A[x-1], A[x]<A[x+1]$ , 称这个元素A[x]为“局部最小”。通过遍历一次数组, 我们可以很容易在 $O(n)$的时间复杂度下找到一个局部最小值, 


1. 分析该问题, 设计一个算法在$O(logn)$的时间复杂度下找到一个局部最小(返回数值), 要求：分析算法设计思路, 并且使用python编程实现
2. \* 设计算法找出所有局部最小值, 分析算法设计思路, 并使用python编程实现

e.g.:
---
A=[9, 3, 7, 2, 1, 4, 5 ] 时,  局部最小元素为 3, 1


数组元素互不相同，并且满足 $A[1]>A[2]$、$A[n-1]<A[n]$。因此从左端看，数组一开始是下降趋势；从右端看，数组最后是上升趋势。只要中间某处由“下降”变成“上升”，就会出现局部最小。

可以用二分查找：

1. 令搜索区间为中间位置，即下标 `1` 到 `n-2`；
2. 取中点 `mid`；
3. 如果 `A[mid]` 同时小于左右邻居，直接返回；
4. 如果 `A[mid] > A[mid-1]`，说明左侧存在更小元素，局部最小一定在左侧；
5. 否则 `A[mid] > A[mid+1]`，局部最小一定在右侧。

每次搜索区间减半，因此时间复杂度为 $O(\log n)$。  
额外空间复杂度为 $O(1)$。  


````python
def find_one_local_min(A):
    """在满足条件的不同整数数组中，返回一个局部最小值。"""
    n = len(A)
    if n < 3:
        raise ValueError("数组长度至少为 3")
    if not (A[0] > A[1] and A[-2] < A[-1]):
        raise ValueError("输入不满足 A[0] > A[1] 且 A[-2] < A[-1]")

    left, right = 1, n - 2
    while left <= right:
        mid = (left + right) // 2
        if A[mid] < A[mid - 1] and A[mid] < A[mid + 1]:
            return A[mid]
        elif A[mid] > A[mid - 1]:
            right = mid - 1
        else:
            left = mid + 1

    raise RuntimeError("未找到局部最小值，理论上不应发生")

A = [9, 3, 7, 2, 1, 4, 5]
print("A =", A)
print("找到的一个局部最小值：", find_one_local_min(A))
print("算法时间复杂度：O(log n)，额外空间复杂度：O(1)")
````

## 问题 5

给定包含n个不同数字的一组数, 寻找一种基于比较的算法在这组数中找到k个最小的数字, 并按顺序输出它们。

1. 将n个数先进行排序, 然后按顺序输出最小的k个数。要求：选择合适的排序算法实现上述操作, 计算算法时间复杂度, 并使用python编程实现。
2. 建立一个包含这n个数的堆（heap）, 并且调用 k 次Extract-min 按顺序输出最小的k个数。使用往空堆中不断插入元素的方法建立堆, 分析这种方法建堆的时间复杂度, 并使用python编程实现
3. \* 假设数组中包含的数据总数目超过了计算机的存储能力, 请设计一个算法, 找到这堆数据的前k小的数值, 计算时间复杂度, 并使用python实现该算法, 假设计算机一定能存储k个数据。

e.g.：
---
数组arr=[5,4,3,2,6,1,88,33,22,107] 的前3个最小数据为：1, 2, 3


1：先排序，再输出前 $k$ 个

使用归并排序。归并排序是基于比较的排序算法，时间复杂度为 $O(n\log n)$。排序完成后输出前 $k$ 个数，额外需要 $O(k)$ 时间。

总时间复杂度：$O(n\log n+k)=O(n\log n)$。  
额外空间复杂度：$O(n)$。

2：从空堆开始逐个插入建堆，再执行 $k$ 次 Extract-min

从空堆开始插入 $n$ 个元素，每次插入复杂度为 $O(\log n)$，所以建堆复杂度为 $O(n\log n)$。之后执行 $k$ 次 Extract-min，每次复杂度为 $O(\log n)$。

总时间复杂度：$O(n\log n+k\log n)$。  
额外空间复杂度：$O(n)$。



````python
def merge_sort(arr):
    if len(arr) <= 1:
        return arr[:]

    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    result = []
    i = j = 0
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])
    return result


def k_smallest_by_sort(arr, k):
    if k < 0 or k > len(arr):
        raise ValueError("k 必须满足 0 <= k <= len(arr)")
    return merge_sort(arr)[:k]


class MinHeap:
    def __init__(self):
        self.data = []

    def push(self, x):
        self.data.append(x)
        i = len(self.data) - 1
        while i > 0:
            parent = (i - 1) // 2
            if self.data[parent] <= self.data[i]:
                break
            self.data[parent], self.data[i] = self.data[i], self.data[parent]
            i = parent

    def extract_min(self):
        if not self.data:
            raise IndexError("heap is empty")
        ans = self.data[0]
        last = self.data.pop()
        if self.data:
            self.data[0] = last
            self._sift_down(0)
        return ans

    def _sift_down(self, i):
        n = len(self.data)
        while True:
            left = 2 * i + 1
            right = 2 * i + 2
            smallest = i
            if left < n and self.data[left] < self.data[smallest]:
                smallest = left
            if right < n and self.data[right] < self.data[smallest]:
                smallest = right
            if smallest == i:
                break
            self.data[i], self.data[smallest] = self.data[smallest], self.data[i]
            i = smallest


def k_smallest_by_heap(arr, k):
    if k < 0 or k > len(arr):
        raise ValueError("k 必须满足 0 <= k <= len(arr)")
    heap = MinHeap()
    for x in arr:       # 从空堆开始逐个插入
        heap.push(x)
    return [heap.extract_min() for _ in range(k)]

arr = [5, 4, 3, 2, 6, 1, 88, 33, 22, 107]
k = 3
print("arr =", arr)
print("k =", k)
print("方法1：排序后取前 k 个：", k_smallest_by_sort(arr, k))
print("方法2：逐个插入建堆后 Extract-min k 次：", k_smallest_by_heap(arr, k))
print("方法1时间复杂度：O(n log n)")
print("方法2时间复杂度：O(n log n + k log n)")
````

## 问题 6

> 星球碰撞问题：现有n个星球，在同一条直线上运行，如数组A所示，元素的绝对值表示星球的质量，负数表示星球自右向左运动，正数表示星球自左向右运动，当两个星球相撞的时候，质量小的会消失，大的保持不变，**质量相同的两个星球碰撞后自右向左运动的星球消失，自左向右的星球保持不变**，假设所有星球的速度大小相同。
>
> $ A=[23,-8, 9, -3, -7, 9, -23, 22] $

请设计一个算法模拟星球的运行情况，输出最终的星球存续情况（输出一个数组），分析算法设计思路，计算时间复杂度，并基于python编程实现。

e.g.
---
1.  输入： A=[-3,-6,2,8, 5,-8,9,-2,1]， 输出：[-3, -6, 2, 8, 9, 1]

2. 输入：A=[23,-8, 9, -3, -7, 9, -23, 22], 输出：[23, 22]


使用栈模拟星球碰撞。

只有当栈顶星球向右运动，即为正数，而当前星球向左运动，即为负数时，二者才可能相撞。比较二者质量：

1. 若栈顶质量小，则栈顶星球消失，当前星球继续向左，可能继续和新的栈顶碰撞；
2. 若栈顶质量大，则当前星球消失；
3. 若质量相同，按照题目规则，自右向左运动的星球消失，自左向右运动的星球保留，所以当前负数星球消失，栈顶正数星球保留。

每个星球最多入栈一次、出栈一次，因此时间复杂度为 $O(n)$。  
额外空间复杂度为 $O(n)$。


````python
def planet_collision(A):
    stack = []

    for x in A:
        alive = True
        while alive and stack and stack[-1] > 0 and x < 0:
            top_mass = abs(stack[-1])
            cur_mass = abs(x)

            if top_mass < cur_mass:
                stack.pop()      # 栈顶星球消失，当前星球继续比较
            elif top_mass == cur_mass:
                alive = False    # 当前负数星球消失，栈顶正数星球保留
            else:
                alive = False    # 当前星球消失

        if alive:
            stack.append(x)

    return stack

examples = [
    [-3, -6, 2, 8, 5, -8, 9, -2, 1],
    [23, -8, 9, -3, -7, 9, -23, 22],
]

for A in examples:
    print("输入：", A)
    print("输出：", planet_collision(A))
    print()

print("算法时间复杂度：O(n)，额外空间复杂度：O(n)")
````

## 问题 7

> 0-1背包问题：现在有4块大理石，每个大理石对应的重量和价值使用一个元组表示, 即（重量，价值），4块大理石的重量和价值为：[(5,10), (4,40), (6,30), (3, 50)], 假设你有一辆最大承重为9的小推车，请问使用这个小推车装大理石的最大价值为多少

请设计一个算法解决该问题，分析算法设计思路，计算时间复杂度，并基于python编程实现


这是 0-1 背包问题。设 `dp[i][w]` 表示只考虑前 `i` 个大理石、容量为 `w` 时能获得的最大价值。

对第 `i` 个大理石，有两种选择：

1. 不装入：价值为 `dp[i-1][w]`；
2. 若重量允许，则装入：价值为 `dp[i-1][w-weight]+value`。

状态转移为：

$$dp[i][w]=\max(dp[i-1][w],\ dp[i-1][w-weight_i]+value_i)$$

其中第二项只有在 $w\ge weight_i$ 时可选。

时间复杂度：$O(nW)$，其中 $n$ 是物品数量，$W$ 是背包容量。  
空间复杂度：$O(nW)$。本题中 $n=4,W=9$。


````python
def knapsack_01(items, capacity):
    """items: [(weight, value), ...]，返回最大价值和选择的物品下标。"""
    n = len(items)
    dp = [[0] * (capacity + 1) for _ in range(n + 1)]

    for i in range(1, n + 1):
        weight, value = items[i - 1]
        for w in range(capacity + 1):
            dp[i][w] = dp[i - 1][w]
            if w >= weight:
                dp[i][w] = max(dp[i][w], dp[i - 1][w - weight] + value)

    # 反向恢复选择了哪些物品
    selected = []
    w = capacity
    for i in range(n, 0, -1):
        if dp[i][w] != dp[i - 1][w]:
            selected.append(i)  # 使用 1-based 编号
            w -= items[i - 1][0]
    selected.reverse()

    return dp[n][capacity], selected

items = [(5, 10), (4, 40), (6, 30), (3, 50)]
capacity = 9
max_value, selected = knapsack_01(items, capacity)
print("大理石列表 (重量, 价值)：", items)
print("小推车最大承重：", capacity)
print("最大价值：", max_value)
print("选择的大理石编号：", selected)
print("选择的大理石：", [items[i - 1] for i in selected])
print("算法时间复杂度：O(nW)，空间复杂度：O(nW)")
````

解释：2 不能在 3 的右子树中，因为 2 < 3 。交换 2 和 3 使二叉搜索树有效。


二叉搜索树的中序遍历结果应该是递增序列。若恰好两个节点值被交换，则中序遍历序列中会出现逆序。

算法步骤：

1. 对树进行中序遍历；
2. 记录上一个访问节点 `prev`；
3. 若发现 `prev.val > cur.val`，说明出现逆序：
   - 第一次出现逆序时，错误节点之一是 `prev`；
   - 每次出现逆序时，另一个错误节点更新为 `cur`；
4. 遍历结束后交换这两个节点的值。

该方法不改变树结构，只交换节点值。

时间复杂度：$O(n)$。  
递归栈空间复杂度：$O(h)$，其中 $h$ 是树高。


````python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    """由层序列表构造二叉树，None 表示空节点。"""
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    q = deque([root])
    i = 1
    while q and i < len(values):
        node = q.popleft()
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            q.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            q.append(node.right)
        i += 1
    return root


def tree_to_list(root):
    """将二叉树转回层序列表，并去掉末尾多余的 None。"""
    if not root:
        return []
    result = []
    q = deque([root])
    while q:
        node = q.popleft()
        if node is None:
            result.append(None)
        else:
            result.append(node.val)
            q.append(node.left)
            q.append(node.right)

    while result and result[-1] is None:
        result.pop()
    return result


def recover_bst(root):
    first = second = prev = None

    def inorder(node):
        nonlocal first, second, prev
        if not node:
            return
        inorder(node.left)

        if prev and prev.val > node.val:
            if first is None:
                first = prev
            second = node
        prev = node

        inorder(node.right)

    inorder(root)
    if first and second:
        first.val, second.val = second.val, first.val
    return root

examples = [
    [1, 3, None, None, 2],
    [3, 1, 4, None, None, 2],
]

for values in examples:
    root = build_tree(values)
    recovered = recover_bst(root)
    print("输入：", values)
    print("恢复后：", tree_to_list(recovered))
    print()

print("算法时间复杂度：O(n)，递归栈空间复杂度：O(h)")
````

设计一个算法计算从根节点到叶节点生成的所有数字之和 ，分析算法设计思路，计算时间复杂度，并基于python编程实现

叶节点:是指没有子节点的节点。

 输入：root = [1,2,3], 输出：25

2.

 <div style="text-align:center;width:200px;"> <img alt="figure" src="./fig/hw3q10-2.jpg"> </div>

 输入：root = [4,9,0,5,1], 输出：1026

解释：从根到叶子节点路径 4->9->5 代表数字 495，4->9->1 代表数字 491，4->0 代表数字 40。因此和为 495 + 491 + 40 = 1026 。


用深度优先搜索从根节点走到叶节点，并在递归过程中维护当前路径形成的数字。

若当前路径之前形成的数字是 `cur`，当前节点值是 `node.val`，则加入当前节点后的数字为：

$$cur \times 10 + node.val$$

当访问到叶节点时，将这个数字加入总和。

每个节点只访问一次，因此时间复杂度为 $O(n)$。  
递归栈空间复杂度为 $O(h)$，其中 $h$ 是树高。


````python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right


def build_tree(values):
    if not values or values[0] is None:
        return None

    root = TreeNode(values[0])
    q = deque([root])
    i = 1
    while q and i < len(values):
        node = q.popleft()
        if i < len(values) and values[i] is not None:
            node.left = TreeNode(values[i])
            q.append(node.left)
        i += 1
        if i < len(values) and values[i] is not None:
            node.right = TreeNode(values[i])
            q.append(node.right)
        i += 1
    return root


def sum_numbers(root):
    def dfs(node, cur):
        if not node:
            return 0
        cur = cur * 10 + node.val
        if not node.left and not node.right:
            return cur
        return dfs(node.left, cur) + dfs(node.right, cur)

    return dfs(root, 0)

examples = [
    [1, 2, 3],
    [4, 9, 0, 5, 1],
]

for values in examples:
    root = build_tree(values)
    print("输入：", values)
    print("所有根到叶路径数字之和：", sum_numbers(root))
    print()

print("算法时间复杂度：O(n)，递归栈空间复杂度：O(h)")
````
